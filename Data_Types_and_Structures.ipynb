4. Fourth Question
-> A dictionary in Python is a collection of key-value pairs. It stores data in a way that allows fast access to values using unique keys, similar to a real-life dictionary where a word (key) maps to a definition (value).

How data is stored internally.
I. Hashing the Key:

Python calculates a hash value using the built-in hash() function.

This hash determines where to store the value in memory.

II. Storing the Key-Value Pair:

The hash points to a bucket (slot) in memory.

The actual key:value pair is stored there.

III. Handling Collisions:

If two keys hash to the same slot (collision), Python uses techniques like open addressing or chaining to store them correctly.
